```python
import pandas as pd
import numpy as np
# 读取CSV文件
df = pd.read_csv('/Users/suxinwan/Documents/sxw-scripts/DB_test/result_data.csv', delimiter=',')

# 初始化一个空的group列
df['group'] = None

# 遍历每个基因列
for gene, group_df in df.groupby('gene_column'):
    # 首行标为 'high'
    group_df.iloc[0, group_df.columns.get_loc('group')] = 'high'
    # 初始化前一时间数据为第一行的时间数据
    prev_time_data = group_df.iloc[0]['time_data']
    # 标记变量，用于判断是否遇到第一个下降点
    first_decline = False
    for i in range(1, len(group_df)):
        current_time_data = group_df.iloc[i]['time_data']
        # 如果当前时间小于前一时间且尚未遇到下降点，则标记为 'low' 并设置下降点标记
        if current_time_data < prev_time_data and not first_decline:
            group_df.iloc[i, group_df.columns.get_loc('group')] = 'low'
            first_decline = True
        # 如果已经遇到下降点，则继续标记为 'low'
        elif first_decline:
            group_df.iloc[i, group_df.columns.get_loc('group')] = 'low'
        # 否则，标记为 'high'
        else:
            group_df.iloc[i, group_df.columns.get_loc('group')] = 'high'
        prev_time_data = current_time_data
    # 更新主DataFrame
    df.update(group_df)

# 输出带有group列的结果
output_file_path = '/Users/suxinwan/Desktop/DB_files/tools_clinical/surv_ccrcc.csv'
with open(output_file_path, 'w') as f:
    df.to_csv(f, index=False)
```

```python
import pandas as pd
import glob

# 获取所有符合surv_*.csv模式的文件路径
file_paths = glob.glob('surv_*.csv')
dfs = []

# 遍历每个文件路径
for file_path in file_paths:
    # 读取CSV文件
    df = pd.read_csv(file_path, delimiter=',')
    
    # 获取文件名中*的内容，用于区分不同文件
    file_identifier = file_path.split('_')[-1].split('.')[0]
    
    # 将时间数据和生存数据合并成字符串，并创建一个新列
    df['time_surv_strings'] = df['time_data'].astype(str) + ',' + df['surv_data'].astype(str)
    
    # 定义一个函数来处理每个基因组
    def process_gene_group(group):
        # 获取基因名称
        gene_name = group['gene_column'].iloc[0]
        # 获取p值
        p_value = group['p_value'].iloc[0]
        
        # 初始化结果列表
        results = []
        
        # 遍历每个状态（高或低）
        for status in group['group'].unique():
            # 检查当前状态是否为高或低
            if status == 'high':
                # 获取高状态下的时间和生存数据，并合并成字符串
                high_time_surv_strings = ';'.join(group[group['group'] == 'high']['time_surv_strings'])
                # 创建一行结果
                result = pd.Series([gene_name, p_value, high_time_surv_strings, status, file_identifier],
                                   index=['gene_name', 'p_value', 'time_surv_strings', 'group_status', 'file_identifier'])
                # 将结果添加到结果列表中
                results.append(result)
            elif status == 'low':
                # 获取低状态下的时间和生存数据，并合并成字符串
                low_time_surv_strings = ';'.join(group[group['group'] == 'low']['time_surv_strings'])
                # 创建一行结果
                result = pd.Series([gene_name, p_value, low_time_surv_strings, status, file_identifier],
                                   index=['gene_name', 'p_value', 'time_surv_strings', 'group_status', 'file_identifier'])
                # 将结果添加到结果列表中
                results.append(result)
        
        # 返回结果列表
        return pd.DataFrame(results)

    # 对DataFrame按基因列进行分组，并应用处理函数
    result_df = df.groupby('gene_column').apply(process_gene_group).reset_index(drop=True)
    
    # 将处理后的DataFrame添加到列表中
    dfs.append(result_df)

# 合并所有DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# 将结果写入文件
output_file_path = f'/Users/suxinwan/Desktop/DB_files/tools_clinical/surv_formatted.csv'
merged_df.to_csv(output_file_path, index=False)
    
print(f"文件 {output_file_path} 已成功写入.")

print("所有文件处理完成.")
```

```python
import pandas as pd

# 读取数据
data1 = pd.read_csv('/Users/suxinwan/Desktop/cptac_reanalysis/ccRCC/results/04co_exp/de_p_exp.csv')
data2 = pd.read_csv('/Users/suxinwan/Desktop/cptac_reanalysis/ccRCC/results/04co_exp/de_r_exp.csv')

# 将数据集按照case_submitter_id进行排序
data1_sorted = data1.sort_values(by='case_submitter_id')
data2_sorted = data2.sort_values(by='case_submitter_id')

# 获取两个数据集的列名
columns = data1.columns.tolist()[1:]
# 创建一个空列表来存储合并后的数据
merged_data = []
# 处理每个列
for column in columns:
    merged_column_data = [f"{column}\t"]
    
    first_value_added = False
    
    for (id1, value1), (id2, value2) in zip(data1_sorted[['case_submitter_id', column]].values, data2_sorted[['case_submitter_id', column]].values):
        assert id1 == id2  # 确保两个数据集的ID匹配
        
        # 在第一个value之后的所有value前面添加分号
        separator = ';' if first_value_added else ''
        merged_column_data.append(f"{separator}{value1},{value2}")
        first_value_added = True

    merged_data.append(''.join(merged_column_data))

# 将合并后的数据按行保存到文件
with open('merged_values.txt', 'w') as f:
    for line in merged_data:
        f.write(f"{line}\n")
```
